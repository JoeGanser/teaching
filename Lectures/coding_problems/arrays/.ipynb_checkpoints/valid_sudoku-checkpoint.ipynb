```python
board = board2
for i in range(0,9):
    row = [r for r in board[i]]# if r!='.']
    column = [row[i] for row in board]# if row[i]!='.']
    if len(set(row))!=len(row) or len(set(column))!=len(column):
        print(i)
        print(row)
        print(column)
        print(False)
```

```python
m = 0 #rows
while m<9:
    n = 0
    for i in range(m,m+3):
        l = []
        for j in range(n,n+3):
            _ = board[i][j]
            if _!='.':
                l.append(_)
            n+=1
            if len(set(l))!=len(l):
                print(False)
                print(l)
        m+=3
```
