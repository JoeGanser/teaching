```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

L = ['minke_whale.n.01','orca.n.01','novel.n.01','right_whale.n.01','tortoise.n.01']

tuples = []
for l in L:
    left = wn.synset(l)
    right = wn.synset('right_whale.n.01')
    distance = left.path_similarity(right)
    tuples.append((left,right,distance))

for hypernym, hyponym,distance in tuples:
    G.add_edge(hypernym,hyponym,distance=distance)
    
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_size=1000,node_color="red",font_size=10,font_color="black",arrowsize=20)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from itertools import permutations

G = nx.DiGraph()

L = ['minke_whale.n.01','orca.n.01','novel.n.01','right_whale.n.01','tortoise.n.01']
perms = permutations(L,2)

tuples = []
for l,r in perms:
    left = wn.synset(l)
    right = wn.synset(r)
    distance = left.path_similarity(right)
    tuples.append((l,r,distance))

for hypernym, hyponym,distance in tuples:
    G.add_edge(hypernym,hyponym,distance=distance)
    
pos = nx.spring_layout(G)
nx.draw(G,pos,with_labels=True,node_size=1000,node_color="red",font_size=10,font_color="black",arrowsize=20)
plt.show()
```
